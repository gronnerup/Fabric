```python
from sempy.fabric import FabricRestClient
try:
    fabric_token = notebookutils.mssparkutils.credentials.getToken("https://api.fabric.microsoft.com")

    class ServicePrincipalTokenProvider:
        def __call__(self):
            return fabric_token

    token_credential = ServicePrincipalTokenProvider()

    client = FabricRestClient(token_credential)
    response = client.post(f"/v1/workspaces/{spark.conf.get('trident.workspace.id')}/items/{notebook_id}/jobs/instances?jobType=RunNotebook")
    print(response)
except:
    print("ERROR")
```

```python
import requests

tenant_id = notebookutils.credentials.getSecret('https://kv-peerinsights-dev.vault.azure.net/', 'Tenant-ID')
client_id = notebookutils.credentials.getSecret('https://kv-peerinsights-dev.vault.azure.net/', 'SPN-Client-ID')
client_secret = notebookutils.credentials.getSecret('https://kv-peerinsights-dev.vault.azure.net/', 'SPN-Client-Secret')

request_access_token_uri = f"https://login.microsoftonline.com/{tenant_id}/oauth2/token"
payload = {
    'grant_type': 'client_credentials',
    'client_id': client_id,
    'client_secret': client_secret,
    'resource': 'https://api.fabric.microsoft.com'
}

response = requests.post(request_access_token_uri, data=payload,headers={'Content-Type': 'application/x-www-form-urlencoded'})    
response.raise_for_status()

access_token = response.json().get('access_token')
print(access_token)

headers = {
    "Authorization": f"Bearer {access_token}",
    "Content-Type": "application/json"
}

try:
    response = requests.post(f"https://api.fabric.microsoft.com/v1/workspaces/{spark.conf.get('trident.workspace.id')}/items/{notebook_id}/jobs/instances?jobType=RunNotebook", headers=headers)
    print(response)
except:
    print("Failed!")
```
