```python
import jwt, json
import sempy.fabric as fabric

# Color codes
cdefault = '\033[0m'
cdefault_bold = '\033[1m'
cred = '\033[91m'
cred_bold = '\033[1;91m'
cyellow = '\033[33m'
cyellow_bold = '\033[1;33m'
cgreen = '\033[32m'
cgreen_bold = '\033[1;32m'
cblue_bold = '\033[1;34m'

def print_header(value):
    print("")
    print(f"{cblue_bold}####################################################################################{cdefault}")
    print(f"{cblue_bold}# {value.center(80)} #{cdefault}")
    print(f"{cblue_bold}####################################################################################{cdefault}")
```

```python
token = notebookutils.mssparkutils.credentials.getToken("https://api.fabric.microsoft.com")
decoded = jwt.decode(token, options={"verify_signature": False})

print_header("MSSPARKUTIL ENVIRONMENT VARIABLES ")
try:
    print(f"Workspace Name: {mssparkutils.env.getWorkspaceName()}")
    print(f"User Id: {mssparkutils.env.getUserId()}")
    print(f"User Name: {mssparkutils.env.getUserName()}")
except:
    print(f"{cred}Failed to get MSSparkUtil environment variables!{cdefault}")

print_header("NOTEBOOKUTILS RUNTIME CONTEXT PROPERTIES")
try:
    print(f"Workspace Name: {notebookutils.runtime.context.get('currentWorkspaceName')}")
    print(f"User Name: {notebookutils.runtime.context.get('userName')}")
except:
    print(f"{cred}Failed to get runtime context properties using NotebookUtils!{cdefault}")

print_header("SPARK CONFIG PROPERTIES")
try:
    print(f"Workspace Id: {spark.conf.get('trident.workspace.id')}")
    print(f"Workspace Name: {spark.conf.get('trident.workspace.name')}")
except:
    print(f"{cred}Failed to get Spark config properties!{cdefault}")
    

print_header("DECODED JWT TOKEN")
print(json.dumps(decoded, indent=4, ensure_ascii=False))
```

```python
from azure.identity import ClientSecretCredential
from sempy.fabric import FabricRestClient

tenant_id = notebookutils.credentials.getSecret('https://kv-peerinsights-dev.vault.azure.net/', 'Tenant-ID')
client_id = notebookutils.credentials.getSecret('https://kv-peerinsights-dev.vault.azure.net/', 'SPN-Client-ID')
client_secret = notebookutils.credentials.getSecret('https://kv-peerinsights-dev.vault.azure.net/', 'SPN-Client-Secret')

class ServicePrincipalTokenProvider:
    def __init__(self, tenant_id, client_id, client_secret):
        self.credential = ClientSecretCredential(
            tenant_id=tenant_id,
            client_id=client_id,
            client_secret=client_secret
        )
        # The scope for the API
        self.scope = "https://analysis.windows.net/powerbi/api/.default"

    def __call__(self):
        token = self.credential.get_token(self.scope)
        decoded = jwt.decode(token.token, options={"verify_signature": False})
        print_header("DECODED JWT TOKEN (FROM SPN KEY VAULT AUTH)")
        print(json.dumps(decoded, indent=4, ensure_ascii=False))
        return token.token

# init service principal credentials
token_provider = ServicePrincipalTokenProvider(
    tenant_id=tenant_id,
    client_id=client_id,
    client_secret=client_secret
)
  
try:
    client = FabricRestClient(token_provider=token_provider)
    response = client.get(f"/v1/workspaces/{spark.conf.get('trident.workspace.id')}")
    print_header("FABRICRESTCLIENT RESPONSE")
    print(response.json())
except:
    print(f"{cred}Failed to get workspace using sempy FabricRestClient")
```
